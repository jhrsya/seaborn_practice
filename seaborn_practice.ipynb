思考一个问题：
sns.replot()需要传入的参数有data（你要表示的数据），x,y是横纵坐标，col是对应每列图，hue是使用不同的颜色表示数据，style是使用不同的格式（所谓格式，就是不同的形状）表示数据，但颜色会是一个样的；size使用不同的大小表示数据

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
print(fmri[:5])
```

```python
sns.relplot(data = fmri, kind = "line", x = "timepoint", y = "signal", col = "region", hue = "event", style = "event")
```

```python
#如果没有hue，那会变成什么样呢？
sns.relplot(data = fmri, kind = "line", x = "timepoint", y = "signal", col = "region", style = "event")
```

很明显，hue就是调节颜色的，而style是调节形状的；至于这里的kind是用于画什么类型的图的,kind = "line"表示画线图

```python
dots = sns.load_dataset("dots")
print(dots[:5])
```

```python
sns.relplot(data = dots, kind = "line", x = "time", y = "firing_rate", col = "align", hue = "choice", size = "coherence", style = "choice", facet_kws = dict(sharex = False))
```

```python
sns.relplot(data = dots, kind = "line", x = "time", y = "firing_rate", col = "align", hue = "choice", size = "coherence", style = "choice")
```

由此可见facet_kws = dict(sharex = False)可以让图变得更集中

```python
sns.lmplot(data = tips, x = "total_bill", y = "tip", col = "time", hue = "smoker")
```
